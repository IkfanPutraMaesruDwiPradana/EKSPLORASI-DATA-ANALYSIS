```python
df.duplicated().sum()
```

#cek data unik di provinsi dan tahun

```python
# Memeriksa nilai unik di kolom 'Provinsi'
unique_Provinsi = df['Provinsi'].unique()
print("Nilai unik di kolom 'Provinsi':", unique_Provinsi)

# Memeriksa nilai unik di kolom 'Tahun'
unique_Tahun = df['Tahun'].unique()
print("Nilai unik di kolom 'Tahun':", unique_Tahun)
```

#Visualisasikan jumlah nilai unik

```python
# Menghitung jumlah nilai unik di kolom 'Provinsi' dan 'Tahun'
provinsi_counts = df['Provinsi'].value_counts()
tahun_counts = df['Tahun'].value_counts()

# Membuat visualisasi bar chart untuk jumlah nilai unik di kolom 'Provinsi'
plt.figure(figsize=(10, 6))
sns.barplot(x=provinsi_counts.index, y=provinsi_counts.values)
plt.xticks(rotation=90)
plt.title('Jumlah Unik Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah')
plt.show()

# Membuat visualisasi bar chart untuk jumlah nilai unik di kolom 'Tahun'
plt.figure(figsize=(17, 8))
sns.barplot(x=tahun_counts.index, y=tahun_counts.values)
plt.title('Jumlah Unik Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

#create a box plot

```python
# Set ukuran figure agar lebih mudah dibaca
plt.figure(figsize=(12, 8))

# Membuat boxplot untuk setiap kolom numerik berdasarkan kategori Provinsi
sns.boxplot(x='Provinsi', y='Produksi', data=df)
plt.xticks(rotation=90)
plt.title('Boxplot Produksi Berdasarkan Provinsi')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Provinsi', y='Luas Panen', data=df)
plt.xticks(rotation=90)
plt.title('Boxplot Luas Panen Berdasarkan Provinsi')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Provinsi', y='Curah hujan', data=df)
plt.xticks(rotation=90)
plt.title('Boxplot Curah Hujan Berdasarkan Provinsi')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Provinsi', y='Kelembapan', data=df)
plt.xticks(rotation=90)
plt.title('Boxplot Kelembapan Berdasarkan Provinsi')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Provinsi', y='Suhu rata-rata', data=df)
plt.xticks(rotation=90)
plt.title('Boxplot Suhu Rata-Rata Berdasarkan Provinsi')
plt.show()
```

#corelation

```python
# Membuat DataFrame dengan hanya kolom numerik yang relevan
kolom_numerik = ['Produksi', 'Luas Panen', 'Curah hujan', 'Kelembapan', 'Suhu rata-rata']
df_numerik = df[kolom_numerik]

# Menghitung matriks korelasi
corr_matrix = df_numerik.corr()

# Visualisasi korelasi menggunakan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
